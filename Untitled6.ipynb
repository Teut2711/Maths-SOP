```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
plt.plot(list(range(1, 10000, 2)), list(map(lambda x:3*x + 1, range(1, 10000, 2)) ))
```

```python
fig = go.Figure(data=go.Scatter(x=list(range(1, 10000, 2)), y=list(map(lambda x:(3*x + 1)//2, range(1, 10000, 2)) ), mode='markers'))
```

```python
fig.show()
```
